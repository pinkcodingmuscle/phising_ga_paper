```python
# imports
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
from src.config import SEED
from src.models import evaluate_final_models
from src.evaluate import evaluate_model
```

```python
# load processed data
with open("../data/processed/uci_plf/uci_plf_split_scaled.pkl", "rb") as f:
    X_train_scaled, X_test_scaled, y_train, y_test = pickle.load(f)

with open("../data/processed/uci_plf/uci_plf_selected_features.pkl", "rb") as f:
    selected_features = pickle.load(f)
```

```python
# build reduced datasets
X_train_final = X_train_scaled[selected_features].copy()
X_test_final = X_test_scaled[selected_features].copy()

print(X_train_final.shape, X_test_final.shape)
```

```python
# evaluate models with GA-selected features
results = evaluate_final_models(X_train_final, y_train)
# results = []
# for name, model in models.items():
  
#   result = evaluate_model(name, model, X_train_final, X_test_final, y_train, y_test)
#   results.append(result)

ga_results_df = pd.DataFrame(results).sort_values(by='Accuracy', ascending=False)
ga_results_df
```

```python
# save GA results
ga_results_df.to_csv("../data/results/uci_plf/uci_plf_ga_results.csv", index=False)
```

```python
# compare baseline and GA results
baseline_results_df = pd.read_csv("../data/results/uci_plf/uci_plf_baseline_results.csv")
comparison_df = ga_results_df.merge(baseline_results_df, on='Model', suffixes=('_GA', '_Baseline'))
comparison_df['Accuracy_Improvement'] = comparison_df['Accuracy_GA'] - comparison_df['Accuracy_Baseline']
comparison_df["F1_Improvement"] = comparison_df['F1_GA'] - comparison_df['F1_Baseline']
comparison_df.sort_values(by='Accuracy_Improvement', ascending=False, inplace=True)
```

```python
# save comparison results
comparison_df.to_csv("../data/results/uci_plf/uci_plf_comparison_results.csv", index=False)
```
