```python
import pandas as pd
import pickle
import sys

sys.path.append("..")

from src.config import SEED
from src.models import get_models
from src.evaluate import evaluate_model
```

```python
# load processed data
with open("../data/processed/uci_plf/uci_plf_split_scaled.pkl", "rb") as f:
    X_train_scaled, X_test_scaled, y_train, y_test = pickle.load(f)
```

```python
# Load models
models = get_models(seed=SEED)
```

```python
# evaluate baseline models
results = []
for name, model in models.items():
  result = evaluate_model(name, model, X_train_scaled, X_test_scaled, y_train, y_test)
  results.append(result)

baseline_results_df = pd.DataFrame(results).sort_values(by='Accuracy', ascending=False)
baseline_results_df
```

```python
# save baseline results
baseline_results_df.to_csv("../data/results/uci_plf/uci_plf_baseline_results.csv", index=False)
```
