```python
import pandas as pd
import pickle
import sys

sys.path.append("..")

from src.config import SEED
from src.models import evaluate_final_models
from src.evaluate import evaluate_model
```

```python
# load processed data
with open("../data/processed/uci_full/uci_full_split_scaled.pkl", "rb") as f:
    X_train_scaled, X_test_scaled, y_train, y_test = pickle.load(f)
```

```python
# Load and tune models on full feature set (baseline - no GA selection)
results_df = evaluate_final_models(X_train_scaled, y_train)
```

```python
# evaluate baseline models
results = []
for _, row in results_df.iterrows():
    result = evaluate_model(row['model'], row['best_estimator'], X_train_scaled, X_test_scaled, y_train, y_test)
    results.append(result)

baseline_results_df = pd.DataFrame(results).sort_values(by='Accuracy', ascending=False)
baseline_results_df
```

```python
# save baseline results
baseline_results_df.to_csv("../data/results/uci_full/uci_full_baseline_results.csv", index=False)
```
