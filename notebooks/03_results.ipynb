```python
# imports
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
from src.config import SEED
from src.models import evaluate_final_models
from src.evaluate import evaluate_model
```

```python
# load processed data
with open("../data/processed/uci_small/uci_small_split_scaled.pkl", "rb") as f:
    X_train_scaled, X_test_scaled, y_train, y_test = pickle.load(f)

with open("../data/processed/uci_small/uci_small_selected_features.pkl", "rb") as f:
    selected_features = pickle.load(f)
```

```python
# build reduced datasets
X_train_final = X_train_scaled[selected_features].copy()
X_test_final = X_test_scaled[selected_features].copy()

print(X_train_final.shape, X_test_final.shape)
```

```python
# evaluate models with GA-selected features
cv_results = evaluate_final_models(X_train_final, y_train)

ga_rows = [
    evaluate_model(row['model'], row['best_estimator'], X_train_final, X_test_final, y_train, y_test)
    for _, row in cv_results.iterrows()
]

ga_results_df = pd.DataFrame(ga_rows).sort_values(by='Accuracy', ascending=False)
ga_results_df
```

```python
# save GA results
ga_results_df.to_csv("../data/results/uci_small/uci_small_ga_results.csv", index=False)
```
